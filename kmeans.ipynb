# Image Segmentation using KMeans

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

### Importing Sample Image

```python
img = cv2.imread("charger.jpeg")
plt.imshow(img)
plt.show()
```

### Color Correction

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

### Linear Array Conversion

```python
img.shape
```

```python
all_pixels = img.reshape(-1,3)
all_pixels.shape
```

### Dominant Color Extraction

```python
km = KMeans(4)
```

```python
km.fit(all_pixels)
```

```python
dominant_colors = km.cluster_centers_
```

```python
dominant_colors
```

```python
dominant_colors = np.array(dominant_colors, dtype = "int")
dominant_colors
```

### Plotting Color Swatches

```python
i = 1
plt.figure(figsize=(10,5))
colors = []
for color in dominant_colors:
    plt.subplot(1,4,i)
    plt.axis("off")
    i+=1
    
    colors.append(color)
    
    a = np.zeros((100,100,3), dtype=int)
    a[:,:,:]=color
    plt.imshow(a)
```

### Segmentation

```python
new_img = np.zeros(all_pixels.shape, dtype="int")
new_img.shape
```

```python
colors
```

```python
km.labels_
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix] = colors[km.labels_[ix]]
```

```python
new_img = new_img.reshape(img.shape)
```

```python
plt.imshow(new_img)
plt.show()
```
